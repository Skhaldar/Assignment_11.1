```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline

url = 'https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv'
titanic = pd.read_csv(url)
titanic.info
```

```python
titanic.head()
```

```python
titanic.groupby("sex").sum()
```

```python
plt.style.use('ggplot')

sex1 = 'female', 'male'
number = [1004, 2000]
pie_color = ("red", "green")
fig, ax = plt.subplots()
ax.pie(number, labels=sex1, autopct='%1.1f%%', colors=pie_color, startangle=90)
ax.axis('equal')
ax.set_title("No. of Females and Males\n", fontsize=24)
plt.show()
```

```python
titanic1 = titanic.dropna(subset=['sex'])
mapping = {'male' : 'blue', 'female' : 'red'}
plt.scatter(titanic1['age'], titanic1['fare'], alpha=0.5, c=titanic1['sex'].map(mapping))
plt.show()
```
